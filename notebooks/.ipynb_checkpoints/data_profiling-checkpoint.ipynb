```python
import os
import pandas as pd

# Path configurations
RAW_DIR = os.path.join("..", "data", "raw_source")

# Mapping the source files to our sector-generic naming structure
file_mapping = {
    "blinkit_orders.csv": "qcommerce_orders",
    "blinkit_order_items.csv": "qcommerce_order_items",
    "blinkit_customers.csv": "qcommerce_customers",
    "blinkit_products.csv": "qcommerce_products",
    "blinkit_inventoryNew.csv": "qcommerce_inventory",
    "blinkit_customer_feedback.csv": "qcommerce_feedback",
    "blinkit_marketing_performance.csv": "qcommerce_marketing"
}

def audit_raw_datasets():
    print(f"[#] Auditing files located in: {os.path.abspath(RAW_DIR)}\n")
    
    for src_file, target_name in file_mapping.items():
        full_path = os.path.join(RAW_DIR, src_file)
        
        if not os.path.exists(full_path):
            print(f"[!] Warning: Missing source file -> {src_file}")
            continue
            
        # Read a preview block of data to inspect structure safely
        df = pd.read_csv(full_path, nrows=100)
        
        print("=" * 60)
        print(f"🎯 Target Interface: {target_name} (from {src_file})")
        print(f"📈 Total columns detected: {len(df.columns)}")
        print("=" * 60)
        
        # Build out metadata structural matrix
        meta_df = pd.DataFrame({
            "Column Name": df.columns,
            "Inferred DataType": df.dtypes.values,
            "Sample Value": df.iloc[0].values
        })
        
        print(meta_df.to_string(index=False))
        print("\n")

if __name__ == "__main__":
    audit_raw_datasets()
```
